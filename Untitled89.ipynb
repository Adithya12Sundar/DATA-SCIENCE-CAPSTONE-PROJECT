```python
class yutayuta(Dataset):
  def __init__(self,df):
    self.feature =torch.tensor(df.drop(['FISH_MMI_COND'] , axis = 1).values , dtype= torch.float32)
    self.target = torch.tensor(df['FISH_MMI_COND'].values, dtype = torch.float32)
  def __getitem__(self, X):
    return self.feature[X], self.target[X]
  def __len__(self):
    return len(self.feature)
```

```python
class offo(nn.Module):
  def __init__(self):
    super(offo, self).__init__()
    self.yo1 = nn.Linear(106,64)
    self.rk = nn.ReLU()
    self.yo2 = nn.Linear(64,48)
    self.rk = nn.ReLU()
    self.yo3 = nn.Linear(48,32)
    self.rk = nn.ReLU()


    self.yo5 = nn.Linear(32, 24)
    self.rk = nn.ReLU()
    self.yo6 = nn.Linear(24,18)
    self.rk = nn.ReLU()
    self.puu = nn.Linear(18,12)
    self.rk4 = nn.ReLU()
    self.op0 = nn.Linear(12,3)
  def forward(self , x):
    x = x
    x = self.yo1(x)
    x = self.rk(x)
    x = self.yo2(x)
    x = self.rk(x)
    x = self.yo3(x)
    x = self.yo5(x)
    x = self.rk(x)
    x = self.yo6(x)
    x = self.rk(x)
    x = self.puu(x)
    x = self.rk(x)
    x = self.op0(x)
    y= self.rk(x)
    return y
```
